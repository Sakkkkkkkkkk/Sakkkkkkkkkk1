```python
# 1) Reading csv file
import pandas as pd
df=pd.read_csv('/content/cricket.csv')
df
```

```python
# 2) Names of cricketer and their respective runs
df.loc[:, ['name', 'runs']]
#df[['name', 'runs']]
```

```python
# 3) Find total wickets taken by them
df['wickets'].sum()
```

```python
# 4) Find average of catches taken
df['catches'].mean()
```

```python
# 5) Find the name of wicketkeeper
df.loc[df['stumpings'] == df['stumpings'].max(), 'name'].values[0]

```

```python
# 6) Print the name of the bowler who played the highest number of matches
df.loc[df['wickets'] == df['wickets'].max(), 'name'].values[0]
```

```python
# 7) Find average of all the bowlers
df['bowling_average'] = df['runs'] / df['wickets']
df
```

```python
# 8) Find name of the bowler with least bowling average
df.loc[df['bowling_average'] == df['bowling_average'].min(), 'name'].values[0]
```

```python
# 9) Print information about all players whose catches per match ratio is greater than 0.5 and store it in a csv file
x = df[df['catches'] / df['matches'] > 0.5]
x.to_csv("output", index=False)
```

```python
# 10) Draw the bar chart of player name against their runs
import matplotlib.pyplot as plt
plt.bar(df['name'], df['runs'])
plt.xlabel("Player Names")
plt.ylabel("Runs Scored")
plt.title("Runs Scored by Each Player")
plt.xticks(rotation=90)
plt.show()
```

```python
# 11) Sort and print information about players by ascending order of runs
df.sort_values(by='runs', ascending=True)
```

```python
# 12) Print the names of players whose wickets are greater than matches
x=df[df['wickets'] > df['matches']]['name']
x
```
